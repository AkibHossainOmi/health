```python
import random
import requests
from bs4 import BeautifulSoup
load_male_pageData = requests.get("https://www.prokerala.com/kids/baby-names/?sort_by=name&name_gender%5B%5D=boy&name_religion%5B%5D=muslim&baby_name_meaning=&start_with%5B%5D=a&start_with%5B%5D=b&start_with%5B%5D=c&start_with%5B%5D=d&start_with%5B%5D=e&start_with%5B%5D=f&start_with%5B%5D=g&start_with%5B%5D=h&start_with%5B%5D=i&start_with%5B%5D=j&start_with%5B%5D=k&start_with%5B%5D=l&start_with%5B%5D=m&start_with%5B%5D=n&start_with%5B%5D=o&start_with%5B%5D=p&start_with%5B%5D=q&start_with%5B%5D=r&start_with%5B%5D=s&start_with%5B%5D=t&start_with%5B%5D=u&start_with%5B%5D=v&start_with%5B%5D=w&start_with%5B%5D=x&start_with%5B%5D=y&start_with%5B%5D=z&process=5&page=5")
load_female_pageData = requests.get("https://www.prokerala.com/kids/baby-names/?sort_by=name&name_gender%5B%5D=girl&name_religion%5B%5D=muslim&baby_name_meaning=&start_with%5B%5D=a&start_with%5B%5D=b&start_with%5B%5D=c&start_with%5B%5D=d&start_with%5B%5D=e&start_with%5B%5D=f&start_with%5B%5D=g&start_with%5B%5D=h&start_with%5B%5D=i&start_with%5B%5D=j&start_with%5B%5D=k&start_with%5B%5D=l&start_with%5B%5D=m&start_with%5B%5D=n&start_with%5B%5D=o&start_with%5B%5D=p&start_with%5B%5D=q&start_with%5B%5D=r&start_with%5B%5D=s&start_with%5B%5D=t&start_with%5B%5D=u&start_with%5B%5D=v&start_with%5B%5D=w&start_with%5B%5D=x&start_with%5B%5D=y&start_with%5B%5D=z&process=2&page=9")
assert load_male_pageData.status_code==200 
assert load_female_pageData.status_code==200 
load_male_pageData = BeautifulSoup(load_male_pageData.content, 'lxml')
load_female_pageData = BeautifulSoup(load_female_pageData.content, 'lxml')
```

**Get Male First Name list**

```python
tables=load_male_pageData.select('table')
maleNameList = []
for table in tables:
  table=table.findAll('span',{"class": "boy-names"})
  for name in table:
    maleNameList.append(name.get_text(strip = True))

print(maleNameList)   


```

**Get Female First Name List**

```python
tables=load_female_pageData.select('table')
femaleNameList = []
for table in tables:
  table=table.findAll('span',{"class": "girl-names"})
  for name in table:
    femaleNameList.append(name.get_text(strip = True))
print(femaleNameList)
```

```python
# Fill Random Data
male_last_name = ['khan', 'ali', 'sarkar', 'mia', 'bhuyan', 
                  'islam', 'hasan','ahmed', 'hossain','chowdhury']
female_last_name = ['akhter', 'khatun','begum']
blood_group = ['A-', 'B+', 'AB+', 'O+', 'B+', 'B-']
proffession = ['doctor','teacher']
```

**Generate Male patient**

```python
#generate Male Patient
malePatientlist = []
for first_name in maleNameList:
  patient = dict.fromkeys(['first_name', 'last_name', 'blood_group', 'gender', 'proffession','address','date_of_birth',
                         'contact_number', 'email', 'official_id_type', 'official_id_number',
                         'last_check_time', 'blood_pressure', 'weight', 'height', 'test_name', 'test_type', 
                         'test_time','test_report_deilivery_time','test_result','contact'])
  patient['first_name'] = first_name
  patient['last_name'] = random.choice(male_last_name)
  patient['blood_group'] = random.choice(blood_group)
  patient['proffession'] = random.choice(proffession)
  patient['gender'] = 'male'

  malePatientlist.append(patient)
import json
maleJsonData = json.dumps(malePatientlist,indent=4)
print(maleJsonData)
```

**Generate Female patient**

```python
femalePatientlist = []
for first_name in femaleNameList:
  patient = dict.fromkeys(['first_name', 'last_name', 'blood_group', 'gender', 'proffession','address','date_of_birth',
                         'contact_number', 'email', 'official_id_type', 'official_id_number',
                         'last_check_time', 'blood_pressure', 'weight', 'height', 'test_name', 'test_type', 
                         'test_time','test_report_deilivery_time','test_result','contact'])
  patient['first_name'] = first_name
  patient['last_name'] = random.choice(female_last_name)
  patient['blood_group'] = random.choice(blood_group)
  patient['proffession'] = random.choice(proffession)
  patient['gender'] = 'male'

  femalePatientlist.append(patient)
import json
femaleJsonString = json.dumps(femalePatientlist,indent=4)
print(femaleJsonString)
```
