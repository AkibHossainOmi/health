```python
import requests
from bs4 import BeautifulSoup
import re
import dateutil
```

```python
load = requests.get("https://www.doctorbangladesh.com/bsmmu-pg-hospital-doctor-list-contact")
#load = requests.get("https://www.doctorbangladesh.com/anesthesiologist-dhaka/")

assert load.status_code==200 
```

```python
src = load.content
pageData = BeautifulSoup(src, 'lxml')
```

```python
doctorInfo = pageData.find_all("div", {"class": "info"})
doctorInfo[0]
```

```python

list = []
dict={}
for singleInfo in doctorInfo:
      dict = {}
      dict['name'] = singleInfo.find_all(['a'])[0].get_text(strip = True)
      incomming = ['title', 'specialization', 'work_place']
      for i in range(3):
        dict[incomming[i]] = singleInfo.find_all('li')[i].get_text(strip = True)
      list.append(dict)     

#print(list)
import json
jsonString = json.dumps(list,indent=4)
print(jsonString)
# with open('dr-list.json', 'w') as testwritefile:
#    testwritefile.write(jsonString)
# print(jsonString)
```
